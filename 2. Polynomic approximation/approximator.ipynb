```python
import math
import numpy as np
from typing import Tuple, List
import scipy.linalg as lg
from matplotlib import pylab as plt
```

```python
def input_func(x: float) -> float:
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
```

```python
def get_linear_equation_system(points: List[float]):
    a = []
    b = []
    for x in points:
        a.append([x ** deg for deg in range(len(points))])
        b.append(input_func(x))

    return a, b
```

```python
def solve(points: List[float]) -> List[float]:
    (a, b) = get_linear_equation_system(points)
    return lg.solve(a, b)
```

```python
def polynom(points: List[float], x):
    c = solve(points)
    print(c)
    
    ans = np.zeros(len(x))
    for deg in range(len(c)):
        ans += c[deg] * x ** deg
    return ans
```

```python
%matplotlib inline
x = np.arange(0, 16, 0.1)
```

```python
vec_input_func = np.vectorize(input_func)
y = vec_input_func(x)
plt.plot(x, y)
plt.show()
```

```python
y = polynom([1, 15], x)
plt.plot(x, y)
plt.show()
```

```python
y = polynom([1, 8, 15], x)
plt.plot(x, y)
plt.show()
```

```python
y = polynom([1, 4, 10, 15], x)
plt.plot(x, y)
plt.show()
```
